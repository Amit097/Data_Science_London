```python
from warnings import filterwarnings
filterwarnings('ignore')

import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture  # Clustering the dataset
from sklearn.ensemble import RandomForestClassifier #classification the test data

%matplotlib inline
```

```python
train=pd.read_csv('./Data/train.csv')
test=pd.read_csv('./Data/test.csv')
train_labels=pd.read_csv('./Data/trainLabels.csv')

train1=train.values
test1=test.values
train_labels1=train_labels.values

print('Shape of Train=',train.shape,'\n','Train_labels=',train_labels.shape,'\n','Test=',test.shape)
train.head()
```

```python
#Combine train and test numpy array into one ndarray
data=np.append(train1,test1,axis=0)
print("shape of combine ndarray=",np.shape(data))
```

```python
# clus
inf_value=np.inf

for i in range(1,10):
    gmm =GaussianMixture(n_components=i,covariance_type='full')
    
    gmm.fit(data)
    aic_value=gmm.aic(data)
    if aic_value < inf_value:
        inf_value=aic_value
        print(aic_value)   #lower aic value is good model
        gmm_next=gmm


gmm_next.fit(data)        
train_data=gmm_next.predict_proba(train1)
test_data=gmm_next.predict_proba(test1)
print('Shape of Train=',np.shape(train_data),'\n','Test=',np.shape(test_data))
```

```python
from sklearn.model_selection import GridSearchCV
rfc=RandomForestClassifier()

grid_search_rfc=GridSearchCV(rfc,param_grid=dict(),cv=10,scoring='accuracy').fit(train_data,train_labels1)

rfc1=grid_search_rfc.best_estimator_

rfc1.fit(train_data,train_labels)
print('Score=',rfc1.score(train_data,train_labels))
```

```python
test_labels=rfc1.predict(test_data)
test_labels=pd.DataFrame(test_labels)

test_labels.to_csv('Submission.csv')
```
